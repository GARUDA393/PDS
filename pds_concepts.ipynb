```python
import pandas as pd
import numpy as np

# Given ages
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]

# Convert ages to a pandas DataFrame
df = pd.DataFrame(ages, columns=['Age'])

# Define the bin edges
bin_edges = [20, 31, 42, 53, 63]

# Use pd.cut() to bin the ages
df['Age Group'] = pd.cut(df['Age'], bins=bin_edges, labels=['20-30', '31-41', '42-52', '53-63'])

print(df)
```

# USING MATPLOTLIB IN DATAFRAME AND DATA SERIES

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'A': [1,2,3,4,5],
    'B': [6,7,8,9,10],
    'C': [11,12,13,14,15]
}
df = pd.DataFrame(data)

# Sample Series
sample = [1,2,3,4,5]
series = pd.Series(sample)

# Plot DataFrame
df.plot()
plt.title('Line Plot of DataFrame Columns')
plt.xlabel('Index')
plt.ylabel('Values')
plt.show()


# Plot Series
series.plot()
plt.title('Line Plot of Series')
plt.xlabel('Index')
plt.ylabel('Values')
plt.show()
```

# JOINING OF DATAFRAMES

```python
import pandas as pd

# Create first DataFrame
df1 = pd.DataFrame({
    'key': ['A', 'B', 'C', 'D'],
    'value1': [1, 2, 3, 4]
})

# Create second DataFrame
df2 = pd.DataFrame({
    'key': ['B', 'D', 'E', 'F'],
    'value2': [5, 6, 7, 8]
})

print("DataFrame 1:")
print(df1)

print("\nDataFrame 2:")
print(df2)

# Left Join
left_join = pd.merge(df1, df2, on='key', how='left')
print("\nLeft Join:")
print(left_join)

# Right Join
right_join = pd.merge(df1, df2, on='key', how='right')
print("\nRight Join:")
print(right_join)

# Inner Join
inner_join = pd.merge(df1, df2, on='key', how='inner')
print("\nInner Join:")
print(inner_join)

# Outer Join
outer_join = pd.merge(df1, df2, on='key', how='outer')
print("\nOuter Join:")
print(outer_join)
```
