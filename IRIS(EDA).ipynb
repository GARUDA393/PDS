# IMPORT REQUIRED LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# LOAD THE DATA SET

```python
iris = pd.read_csv('iris.data.csv')
iris.head()
```

```python
iris.info()
iris.describe()
```

# CHECK FOR NULL VALUES

```python
iris.isnull()
```

```python
iris.isnull().sum()
```

# TO VISUALISE MISSING VALUES

```python
sns.heatmap(iris.isnull())
```

# CHECK FOR DUPLICATE VALUES

```python
iris.duplicated()
iris.duplicated().sum()
```

# ERRADICATE DUPLICATE VALUES

```python
iris.drop_duplicates()
```

# CHECK FOR DATA TYPES

```python
iris.dtypes
```

# DATA VISUALISATION

```python
iris.plot()
```

```python
iris.plot(subplots=True)
```

```python
iris['Iris-setosa'].value_counts().plot.pie()
plt.title('Distribution of Iris Species')
plt.ylabel('')
plt.show()
```

```python
sns.countplot(data=iris, x='Iris-setosa')
plt.title('Distribution of Iris Species')
plt.xlabel('Iris-setosa')
plt.ylabel('Count')
plt.show()
```
