```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
data = pd.read_csv('data.csv')
data

#Distribution of Customer Lifetime Value:
plt.figure(figsize=(10, 6))
sns.histplot(data['Customer Lifetime Value'], kde=True)
plt.title('Distribution of Customer Lifetime Value')
plt.xlabel('Customer Lifetime Value')
plt.ylabel('Frequency')
plt.show()

# Income Distribution by Gender:
sns.boxplot(x='Gender', y='Income', data=data)
plt.title('Income Distribution by Gender')
plt.show()

# Monthly Premium Auto vs Total Claim Amount
sns.scatterplot(x='Monthly Premium Auto', y='Total Claim Amount', data=data)
plt.title('Monthly Premium Auto vs Total Claim Amount')
plt.xlabel('Monthly Premium Auto')
plt.ylabel('Total Claim Amount')
plt.show()

# Distribution of Income
plt.figure(figsize=(10, 6))
sns.histplot(data['Income'], kde=True)
plt.title('Distribution of Income')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()

# Relationship between Income and Customer Lifetime Value
sns.scatterplot(x='Income', y='Customer Lifetime Value', data=data)
plt.title('Income vs Customer Lifetime Value')
plt.xlabel('Income')
plt.ylabel('Customer Lifetime Value')
plt.show()
```
