# IMPORT REQUIRED LIBRARIES

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# LOADING THE DATA

```python
data = pd.read_csv("train.csv")
data.head()
```

# DATA PREPROCCESSING

# 1. MISSING VALUES

```python
data.isnull()
data.isnull().sum()
```

# Visualise missing values

```python
sns.heatmap(data.isnull())
```

# Replace missing values

```python
data.replace(np.nan, '0', inplace= True)
data.isnull().sum()
```

# CHECK FOR DUPLICATE VALUES

```python
data.duplicated().sum()
```

# IDENTIFY DATA TYPES

```python
data.dtypes
```

# FILTER THE DATA (FINDING OUTLAYER)

```python
data[data['Pclass']==1].head()
```

# DATA VISUALISATION

```python
sns.countplot(x='Survived', data= data)
```

```python
class_distribution = data['Pclass'].value_counts()
labels = class_distribution.index
sizes = class_distribution.values
plt.pie(sizes, labels=labels)
plt.title('Passenger Class Distribution')
plt.show()
```

```python
sns.lineplot(data= data, x='Age', y='Fare')
plt.title('Average Fare by Age')
plt.xlabel('Age')
plt.ylabel('Average Fare')
plt.show()
```
